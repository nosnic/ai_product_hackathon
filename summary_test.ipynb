```python
import importlib
import YandexGPT_API
importlib.reload(YandexGPT_API)
import pandas as pd
from time import sleep
```

```python
news_cut = pd.read_csv('news_cut.csv').head(50)
news_cut.head()
```

```python
gpt = YandexGPT_API.YandexGPTEmbeddings()
```

```python
sum_texts = []
for text in  news_cut['text']:
    sleep(0.5)
    message = [
        {
        "role": "system",
        "text": "Ты лучший редактор на свете, выведи краткое содержание статьи снизу"
        },
        {
        "role": "user",
        "text": text
        }
    ]
    try:
        answer = gpt.make_request(message, temperature=0.2, max_tokens=300)
    except Exception as e:
        print(f"Error occurred: {e}")
        answer = ''
    sum_texts.append(answer)
```

```python
news_cut['summary'] = sum_texts
news_cut.to_csv('news_summary.csv')
```
